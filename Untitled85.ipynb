```python
# Step 1 — Install dependencies
!pip uninstall -y box2d box2d-py gym gymnasium stable-baselines3
!pip install gymnasium[classic-control]==0.29.1 stable-baselines3==2.2.1 matplotlib numpy pandas seaborn

# Step 2 — Imports
import gymnasium as gym
from gymnasium import spaces
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.callbacks import EvalCallback

# Step 3 — Enhanced CyberKnife Simulation Environment
class CyberKnifeSimEnv(gym.Env):
    def __init__(self, render_mode=None, patient_data=None):
        super().__init__()
        self.action_space = spaces.Box(low=np.array([-0.2, -0.2]), high=np.array([0.2, 0.2]), dtype=np.float32)
        self.observation_space = spaces.Box(low=-1, high=1, shape=(5,), dtype=np.float32)
        self.render_mode = render_mode
        self.patient_data = patient_data
        self.grid_size = 50
        self.dose_map = np.zeros((self.grid_size, self.grid_size))
        self.reset()

    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        if self.patient_data is not None:
            patient = self.patient_data.sample(1).iloc[0]
            self.tumor_pos = np.array([patient["tumor_x"], patient["tumor_y"]])
            self.oar_positions = [np.array([patient["oar1_x"], patient["oar1_y"]]),
                                  np.array([patient["oar2_x"], patient["oar2_y"]])]
        else:
            self.tumor_pos = np.random.uniform(-0.5, 0.5, size=2)
            self.oar_positions = [np.random.uniform(-0.5, 0.5, size=2) for _ in range(2)]

        self.beam_pos = np.random.uniform(-1, 1, size=2)
        self.intensity = np.random.uniform(0.5, 1.0)
        self.total_dose = 0
        self.steps = 0
        self.dose_map = np.zeros((self.grid_size, self.grid_size))
        return np.concatenate([self.beam_pos, self.tumor_pos, [self.intensity]]), {}

    def step(self, action):
        self.beam_pos += action
        self.beam_pos = np.clip(self.beam_pos, -1, 1)

        # Simulate small random tumor motion (e.g., breathing)
        self.tumor_pos += np.random.normal(0, 0.01, size=2)
        self.tumor_pos = np.clip(self.tumor_pos, -0.5, 0.5)

        tumor_dist = np.linalg.norm(self.beam_pos - self.tumor_pos)
        tumor_dose = np.exp(-3 * tumor_dist) * self.intensity
        oar_dose = sum(np.exp(-3 * np.linalg.norm(self.beam_pos - o)) * self.intensity for o in self.oar_positions)

        reward = tumor_dose - 0.6 * oar_dose
        self.total_dose += tumor_dose
        self.steps += 1

        # Update dose heatmap
        bx = int((self.beam_pos[0] + 1) * self.grid_size / 2)
        by = int((self.beam_pos[1] + 1) * self.grid_size / 2)
        if 0 <= bx < self.grid_size and 0 <= by < self.grid_size:
            self.dose_map[bx, by] += tumor_dose

        done = self.steps >= 50
        obs = np.concatenate([self.beam_pos, self.tumor_pos, [self.intensity]])
        info = {"tumor_dose": tumor_dose, "oar_dose": oar_dose, "total_dose": self.total_dose}
        return obs, reward, done, False, info

    def render(self):
        plt.figure(figsize=(5,5))
        sns.heatmap(self.dose_map.T, cmap="YlOrRd", cbar=True)
        plt.scatter((self.tumor_pos[0] + 1) * 25, (self.tumor_pos[1] + 1) * 25, c='red', s=100, label='Tumor')
        for i, o in enumerate(self.oar_positions):
            plt.scatter((o[0] + 1) * 25, (o[1] + 1) * 25, c='orange', s=80, label=('OAR' if i==0 else None))
        plt.title("Dose Heatmap - CyberKnife Simulation")
        plt.legend()
        plt.show()

# Step 4 — Generate synthetic "patient dataset"
def generate_patient_dataset(n=100):
    data = {
        "tumor_x": np.random.uniform(-0.4, 0.4, n),
        "tumor_y": np.random.uniform(-0.4, 0.4, n),
        "oar1_x": np.random.uniform(-0.5, 0.5, n),
        "oar1_y": np.random.uniform(-0.5, 0.5, n),
        "oar2_x": np.random.uniform(-0.5, 0.5, n),
        "oar2_y": np.random.uniform(-0.5, 0.5, n),
    }
    df = pd.DataFrame(data)
    df.to_csv("patient_dataset.csv", index=False)
    return df

patient_data = generate_patient_dataset(200)
print(" Synthetic patient dataset created: patient_dataset.csv")

# Step 5 — Train PPO Agent
def make_env():
    return CyberKnifeSimEnv(patient_data=patient_data)

train_env = DummyVecEnv([make_env])
eval_env = DummyVecEnv([make_env])
eval_callback = EvalCallback(eval_env, best_model_save_path='./cyberknife_model/',
                             log_path='./cyberknife_logs/', eval_freq=2000,
                             deterministic=True, render=False)

model = PPO("MlpPolicy", train_env, verbose=1)
print("Starting PPO training ... (approx. 3 mins)")
model.learn(total_timesteps=20000, callback=eval_callback)
print("Training complete!")

# Step 6 — Visualize dose and beam trajectory
env = CyberKnifeSimEnv(patient_data=patient_data)
obs, _ = env.reset()
positions, tumor_doses, oar_doses = [env.beam_pos.copy()], [], []

for _ in range(50):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, truncated, info = env.step(action)
    positions.append(env.beam_pos.copy())
    tumor_doses.append(info["tumor_dose"])
    oar_doses.append(info["oar_dose"])
    if done or truncated:
        break

positions = np.array(positions)

# Beam Path Visualization
plt.figure(figsize=(5,4))
plt.plot(positions[:,0], positions[:,1], 'bo-', label='Beam Path')
plt.scatter(*env.tumor_pos, c='red', s=120, label='Tumor')
for i,o in enumerate(env.oar_positions):
    plt.scatter(*o, c='orange', s=80, label=('OAR' if i==0 else None))
plt.legend(); plt.title('Beam Trajectory'); plt.show()

# Dose Graphs
plt.figure(figsize=(6,3))
plt.plot(tumor_doses, label='Tumor Dose')
plt.plot(oar_doses, label='OAR Dose')
plt.xlabel("Step"); plt.ylabel("Dose")
plt.legend(); plt.title("Dose Over Time"); plt.show()

# Dose Heatmap
env.render()
```
